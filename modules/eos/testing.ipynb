```python
import eos_functions as ef 
from eos_functions import * 
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
P = np.linspace(10000, 2E16, 1000)
M = 5
U_init= np.linspace(1, 5, 1000)
mu = 1.004
```

```python
temp = ef.temperature_solver(M, mu, U)
```

```python
rho = np.zeros(shape=(len(P), len(temp)))
for i, p in enumerate(P): 
    for j,t in enumerate(temp): 
        rho[i,j] = ef.simple_eos(p, mu, t)
```

```python
fig,ax = plt.subplots(1,1,figsize = (8,6))
ax.contourf(temp, P,np.log10(rho), levels = 5)
cbar = plt.colorbar(cm.ScalarMappable(cmap='viridis'), ax=ax)
cbar.set_label(r'$\rho$', rotation=90, fontsize = 12)
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('temperature', fontsize = 12)
ax.set_ylabel('pressure', fontsize = 12)
plt.savefig('../../output/plots/polytropic_eos_colorplot2.png', dpi=300)
```
