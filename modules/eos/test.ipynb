```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
mu = 1.004
T = np.linspace(0.0001, 1E5, 1000)
P = T
rho = ef.simple_eos(P, T, mu)
radius = 1/(T**2)
```

```python
fig, ax = plt.subplots(2,2, figsize = (8,6))
ax[1,0].plot(P/np.max(P), T/np.max(T))
ax[0,0].plot(P/np.max(P),rho/np.max(rho))
ax[1,1].plot(rho/np.max(rho), T/np.max(T))
ax[1,0].set_xlabel('Pressure (normalized)')
ax[1,1].set_xlabel('Density (normalized)')
ax[1,0].set_ylabel('Temperature (normalized)')
ax[0,0].set_ylabel('Density (normalized)')
ax[0,1].set_visible(False)
```
